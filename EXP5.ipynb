Perform exploratory data analysis (EDA) on a dataset containing student scores in different subjects. Show correlations using a heatmap.
heatmap visualizing correlations between subjects like Math, Science, and English.

Steps for EDA and Heatmap Visualization:
1.Import necessary libraries.
2.Load the dataset into a Pandas DataFrame.
3.Inspect the data (summary statistics, missing values, etc.).
4.Visualize the correlation matrix using a heatmap.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from your CSV file (replace 'your_file.csv' with the actual file path)
df = pd.read_csv('student_marks.csv')  # Replace with your CSV file path

# Inspect the first few rows of the dataset
print(df.head())

# Check for missing values in the dataset
print(df.isnull().sum())

# Select specific columns (Maths, Physics, Chemistry, English, Biology)
selected_columns = ['Maths', 'Physics', 'Chemistry', 'English', 'Biology']
df_selected = df[selected_columns]

# Ensure all selected columns are numeric
df_selected = df_selected.apply(pd.to_numeric, errors='coerce')

# Check for missing values in the selected columns
print(df_selected.isnull().sum())

# Calculate the correlation matrix for the selected numeric columns
correlation_matrix = df_selected.corr()

# Display the correlation matrix
print(correlation_matrix)

# Plotting the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap between Maths, Physics, Chemistry, English, and Biology")
plt.show()
```
