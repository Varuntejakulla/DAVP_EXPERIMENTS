Visualize the frequency distribution of ages in a dataset using a histogram.
A histogram showing age distribution.

steps:

路  Data Setup: We created a dataset with an Age column.
路  Histogram: The plt.hist() function is used to plot the histogram. The number of bins is set to 10, meaning the age range is divided into 10 intervals.
路  Customization: Titles and axis labels are added for clarity.
路  Plot Display: plt.show() is used to display the plot


```python
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset with ages
data = {
    'Age': [23, 45, 36, 50, 29, 38, 41, 55, 42, 33, 25, 31, 47, 60, 29, 34, 41, 52, 46, 28]
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Plot the histogram
plt.figure(figsize=(8, 6))  # Set figure size
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')  # Create histogram

# Add title and labels
plt.title('Age Distribution', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Display the plot
plt.show()
```
