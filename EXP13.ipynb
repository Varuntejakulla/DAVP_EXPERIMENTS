Create an infographic using Matplotlib summarizing the key insights from a dataset about employee performance ratings.A combination of bar charts, pie charts, and annotations summarizing performance ratings.

steps:
1.Bar Chart (Top Left): Displays the number of projects completed by each employee.
2.Pie Chart (Top Right): Shows the distribution of performance ratings as percentages.
3.Bar Chart (Bottom Left): Represents attendance percentages for each employee.
4.Annotations (Bottom Right): Provides a text summary of key insights derived from the data.

```python
import pandas as pd

# Original dataset
data = {
    "Employee": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Performance Rating": ["Excellent", "Good", "Excellent", "Average", "Good"],
    "Projects Completed": [10, 8, 12, 6, 9],
    "Attendance (%)": [95, 88, 90, 85, 92]
}

# Add more data
additional_data = {
    "Employee": ["Frank", "Grace", "Hannah", "Ian", "Jack"],
    "Performance Rating": ["Good", "Excellent", "Average", "Good", "Excellent"],
    "Projects Completed": [7, 11, 5, 8, 13],
    "Attendance (%)": [87, 96, 82, 89, 97]
}

# Combine datasets
data_combined = pd.concat([pd.DataFrame(data), pd.DataFrame(additional_data)], ignore_index=True)

# Save to CSV
data_combined.to_csv("employee_performance.csv", index=False)

print("Dataset saved as 'employee_performance.csv'")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read dataset from CSV
df = pd.read_csv("employee_performance.csv")

# Count of performance ratings for the pie chart
rating_counts = df["Performance Rating"].value_counts()

# Set up the figure and axes
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Employee Performance Infographic", fontsize=16, weight="bold")

# Bar Chart: Projects Completed by Employee
axes[0, 0].bar(df["Employee"], df["Projects Completed"], color='skyblue', edgecolor='black')
axes[0, 0].set_title("Projects Completed by Employee")
axes[0, 0].set_xlabel("Employee")
axes[0, 0].set_ylabel("Projects Completed")

# Pie Chart: Distribution of Performance Ratings
axes[0, 1].pie(
    rating_counts,
    labels=rating_counts.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=["green", "orange", "red"]
)
axes[0, 1].set_title("Performance Ratings Distribution")

# Bar Chart: Attendance Percentage
axes[1, 0].bar(df["Employee"], df["Attendance (%)"], color='lightcoral', edgecolor='black')
axes[1, 0].set_title("Attendance Percentage")
axes[1, 0].set_xlabel("Employee")
axes[1, 0].set_ylabel("Attendance (%)")

# Annotations: Summary Insights
axes[1, 1].axis("off")  # Turn off axis for annotation box
annotations = [
    "Key Insights:",
    "- Employees with 'Excellent' ratings completed the most projects.",
    "- Attendance rates are generally above 85%.",
    "- Good performers completed fewer projects compared to Excellent performers."
]
for i, annotation in enumerate(annotations):
    axes[1, 1].text(0, 1 - i * 0.2, annotation, fontsize=12)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Leave space for the main title

# Show the infographic
plt.show()
```
