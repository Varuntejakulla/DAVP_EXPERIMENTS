
Create a scatterplot to analyze the relationship between advertising expenses and sales from a dataset.
A scatterplot showing the positive or negative


Steps to Create a Scatterplot:
1.Install Required Libraries: Install pandas, seaborn, and matplotlib to handle data and create plots.
2.Load the Dataset:Load your dataset into a pandas DataFrame (ensure your file path is correct).
3.Check for Missing Data: Check for any missing values in your dataset that could affect the analysis.
4.Create a Scatterplot
1.Set the figure size for better visualization.
2.Create a scatterplot to show the relationship between the two variables (e.g., Advertising and Sales).
3.Customize the plot with a title and axis labels.
5.Interpret the Scatterplot:
1.Positive Correlation: If the points move upward as you go right, it indicates a positive correlation (higher advertising leads to higher sales).
2.Negative Correlation: If the points move downward as you go right, it suggests a negative correlation (higher advertising leads to lower sales).
3.No Clear Relationship: If the points are randomly scattered with no clear trend, there is no significant correlation.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (replace 'your_file.csv' with the actual path to your dataset)
df = pd.read_csv('advertising.csv')  # Ensure you have the correct path to your dataset

# Inspect the first few rows of the dataset to ensure it's loaded correctly
print(df.head())

# Calculate the correlation between advertising expenses (TV, Radio, Newspaper) and Sales
correlation_tv_sales = df['TV'].corr(df['Sales'])
correlation_radio_sales = df['Radio'].corr(df['Sales'])
correlation_newspaper_sales = df['Newspaper'].corr(df['Sales'])

# Print the correlation values
print(f"Correlation between TV Advertising and Sales: {correlation_tv_sales}")
print(f"Correlation between Radio Advertising and Sales: {correlation_radio_sales}")
print(f"Correlation between Newspaper Advertising and Sales: {correlation_newspaper_sales}")

# Create scatterplots for each advertising expense (TV, Radio, Newspaper) vs Sales

# Set the figure size
plt.figure(figsize=(12, 8))

# Scatterplot for TV vs Sales
plt.subplot(2, 2, 1)  # Creates a 2x2 grid, position 1
sns.scatterplot(data=df, x='TV', y='Sales', color='blue')
plt.title('TV Advertising vs Sales')
plt.xlabel('TV Advertising Expenses')
plt.ylabel('Sales')

# Scatterplot for Radio vs Sales
plt.subplot(2, 2, 2)  # Creates a 2x2 grid, position 2
sns.scatterplot(data=df, x='Radio', y='Sales', color='green')
plt.title('Radio Advertising vs Sales')
plt.xlabel('Radio Advertising Expenses')
plt.ylabel('Sales')

# Scatterplot for Newspaper vs Sales
plt.subplot(2, 2, 3)  # Creates a 2x2 grid, position 3
sns.scatterplot(data=df, x='Newspaper', y='Sales', color='red')
plt.title('Newspaper Advertising vs Sales')
plt.xlabel('Newspaper Advertising Expenses')
plt.ylabel('Sales')

# Display the plots
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
