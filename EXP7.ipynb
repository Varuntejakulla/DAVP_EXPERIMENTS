Use Pandas to group and aggregate a dataset on city-wise COVID-19 cases. Visualize the data with a bar chart.
A bar chart showing total cases for each city.

Steps:
路  Load the Data: The dataset is loaded from a CSV file into a Pandas DataFrame.
路  Group and Aggregate: The groupby() function is used to group the data by City, and sum() aggregates the number of COVID-19 cases for each city.
路  Sorting: The cities are sorted in descending order based on total cases to make the bar chart clearer.
路  Plotting: matplotlib is used to plot the bar chart. The plt.bar() function is used to create the bars, and axis labels and a title are added for clarity.


```python
import pandas as pd

# Creating the extended dataset
data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'San Diego', 'Dallas', 'Miami', 
             'Philadelphia', 'Atlanta', 'Seattle', 'Denver', 'Boston', 'San Francisco', 'Washington', 
             'Austin', 'Orlando', 'Detroit', 'Las Vegas', 'Minneapolis'],
    'Cases': [150000, 120000, 80000, 70000, 45000, 60000, 55000, 100000, 90000, 75000, 
              65000, 70000, 85000, 105000, 95000, 72000, 50000, 60000, 110000, 75000]
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file in the current directory
df.to_csv('./covid_data.csv', index=False)

# Display the first few rows to verify
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace 'covid_data.csv' with your file path)
df = pd.read_csv('covid_data.csv')

# Inspect the first few rows of the dataset
print(df.head())

# Assuming the dataset has columns: 'City' and 'Cases'
# Group by 'City' and aggregate the total cases using sum
city_cases = df.groupby('City')['Cases'].sum().reset_index()

# Sort the values by total cases in descending order for better visualization
city_cases = city_cases.sort_values(by='Cases', ascending=False)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(city_cases['City'], city_cases['Cases'], color='skyblue')
plt.title('Total COVID-19 Cases by City', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('Total Cases', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate city names for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
